```python
dfm = df
# Outliers
Торговельна_площа = []
for Торговельна_площа in df['Торговельна_площа']:
    if Торговельна_площа > (df['Торговельна_площа'].mean() + 3*df['Торговельна_площа'].std()):
        Торговельна_площа = (dfm['Торговельна_площа'].mean() + 3*df['Торговельна_площа'].std())
    elif Торговельна_площа < (df['Торговельна_площа'].mean() - 3*df['Торговельна_площа'].std()):
        Торговельна_площа = (dfm['Торговельна_площа'].mean() - 3*df['Торговельна_площа'].std())
    Торговельна_площа.append(Торговельна_площа)
df['Торговельна_площа'] = Торговельна_площа

количество_касс = []
for количество_касс in df['количество_касс']:
    if количество_касс > (df['количество_касс'].mean() + 3*df['количество_касс'].std()):
        количество_касс = (dfm['количество_касс'].mean() + 3*df['количество_касс'].std())
    elif количество_касс < (df['количество_касс'].mean() - 3*df['количество_касс'].std()):
        количество_касс = (dfm['количество_касс'].mean() - 3*df['количество_касс'].std())
    количество_касс.append(количество_касс)
df['количество_касс'] = количество_касс

Кількість_відгуків_в_гуглі = []
for Кількість_відгуків_в_гуглі in df['Кількість_відгуків_в_гуглі']:
    if Кількість_відгуків_в_гуглі > (df['Кількість_відгуків_в_гуглі'].mean() + 3*df['Кількість_відгуків_в_гуглі'].std()):
        Кількість_відгуків_в_гуглі = (dfm['Кількість_відгуків_в_гуглі'].mean() + 3*df['Кількість_відгуків_в_гуглі'].std())
    elif Кількість_відгуків_в_гуглі < (df['Кількість_відгуків_в_гуглі'].mean() - 3*df['Кількість_відгуків_в_гуглі'].std()):
        Кількість_відгуків_в_гуглі = (dfm['Кількість_відгуків_в_гуглі'].mean() - 3*df['Кількість_відгуків_в_гуглі'].std())
    Кількість_відгуків_в_гуглі.append(Кількість_відгуків_в_гуглі)
df['Кількість_відгуків_в_гуглі'] = Кількість_відгуків_в_гуглі

количество_чеков = []
for количество_чеков in df['количество_чеков']:
    if количество_чеков > (df['количество_чеков'].mean() + 3*df['количество_чеков'].std()):
        количество_чеков = (dfm['количество_чеков'].mean() + 3*df['количество_чеков'].std())
    elif количество_чеков < (df['количество_чеков'].mean() - 3*df['количество_чеков'].std()):
        количество_чеков = (dfm['количество_чеков'].mean() - 3*df['количество_чеков'].std())
    количество_чеков.append(количество_чеков)
df['количество_чеков'] = количество_чеков
```

```python
# Removing Correlated Data
df = df.drop(['Лікарня'], axis='columns', inplace=False)
df = df.drop(['Поліклініка'], axis='columns', inplace=False)
df = df.drop(['Торговий_центр'], axis='columns', inplace=False)
df = df.drop(['Фастфуд'], axis='columns', inplace=False)
df = df.drop(['Тип'], axis='columns', inplace=False)
```

```python
a = ['Торговельна_площа', 'Кількість_відгуків_в_гуглі', 'Количество_чеков', 'Среднее_количество_единиц_в_чеке','Середній_дохід_населення','Населення_середнє']
```

```python
#linear regression
#Splitting the dataset into the Training set and Test set
X =df[a]
y = df['Количество_касс']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# Fitting Simple Linear Regression to the Training set
f1 = ['Количество_чеков']

from sklearn.linear_model import LinearRegression
sr = LinearRegression().fit(X_train[f1], y_train)

# Getting parameters
sr.coef_, sr.intercept_
```

```python
# Predicting the Test set results
y_pred = sr.predict(X_test[f1])

# Coefficient of determination R^2
sr.score(X_train[f1], y_train).round(3), sr.score(X_test[f1], y_test).round(3)
```

```python
# Visualising the Test set results
plt.scatter(X_test['Количество_чеков'], y_test, color = 'red')
plt.plot(X_test['Количество_чеков'], sr.predict(X_test[f1]), 'bo')
plt.title('Количество_касс vs Количество_чеков (Test set)')
plt.xlabel('Количество_чеков')
plt.ylabel('Количество_касс')
plt.show()
```

```python
def metrics(X_train, X_test, y_train, y_test, y_pred, model):
    RMSE = [mean_squared_error(y_train, model.predict(X_train),squared=False), mean_squared_error(y_test, y_pred, squared=False)]
    R2 = [model.score(X_train, y_train), model.score(X_test, y_test)]
    results = {'RMSE': RMSE, 'R2':R2}
    return pd.DataFrame(results, index=['Train', 'Test'])
```

```python
# Fitting Random Forest to the Training set
rf = RandomForestRegressor(n_estimators = 2, max_leaf_nodes = 25, random_state = 10).fit(X_train, y_train)

# Predicting the Test set results
rf_y_pred = rf.predict(X_test)

# Metrics
metrics(X_train, X_test, y_train, y_test, rf_y_pred, rf).round(2)
```
